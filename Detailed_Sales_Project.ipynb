```python
# Lets find the missing columns 

def missing_col(df):
    missingcol=df.isna().sum()
    missingcoldf=pd.DataFrame(missingcol[missingcol>0])
    return missingcoldf

train_missing=missing_col(train_set)
train_missing   
```

```python
# Now lets calculate the percentage of missing values to decide what we do with those

def Nullperc(df,df_missing):
    df_missing=df_missing.copy()
    df_missing["percentage"]=df_missing[0].apply(lambda x:round((x/df.shape[0])*100,2))
    return df_missing

train_missing=Nullperc(train_set,train_missing)
train_missing
```

```python
test_missing=Nullperc(test_set,test_missing)
test_missing
```

```python
### Lets have a policy that if the feature has above of 50 percent missing values we drop that feature and
###  if it is below the 50 percent we can fill the missing values
from sklearn.impute import SimpleImputer

import pandas as pd
from sklearn.impute import SimpleImputer

def columnselection(df, miscoldf):
    columns_to_drop = miscoldf[miscoldf.percentage > 50].index.tolist()
    impute = miscoldf[miscoldf.percentage <= 50].index.tolist()
    
    # Drop columns instead of rows
    df = df.drop(columns=columns_to_drop, axis=1)
    
    if impute:
        numeric = [col for col in impute if pd.api.types.is_numeric_dtype(df[col])]
        categorical = list(set(impute) - set(numeric))
        
        if numeric:
            num_imputer = SimpleImputer(strategy="median")
            df[numeric] = num_imputer.fit_transform(df[numeric])
        
        if categorical:
            cat_imputer = SimpleImputer(strategy="most_frequent")
            df[categorical] = cat_imputer.fit_transform(df[categorical])
    
    return df

train_set = columnselection(train_set, train_missing)


        
```

```python
### Now we have cleaned up the missing datas we can actually convert object dtypes to categories by encoding them 
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import TargetEncoder
```

```python
def Encoding_features(train,test,label=None,OHC=None,Target_enc=None,target_col=None):
   # 1.Label encoding for comparable categories
    if label:
        for col in label:
            if col in train.columns:
                train[col]=train[col].astype(str).fillna("missing")
                if col in test.columns:
                    test[col]=test[col].astype(str).fillna("missing")

                    encoder=LabelEncoder()

                    train[col]=encoder.fit_transform(train[col])
                    test[col]=encoder.transform(test[col])


  #2.One hot encoding for Nominal features with low cardinaty
    if OHC:
        train=pd.get_dummies(train,columns=OHC,drop_first=True)
        test=pd.get_dummies(test,columns=OHC,drop_first=True)

        missing_cols=set(train.columns)-set(test.columns)

        for col in missing_cols:
            test[col]=0
        test=test[train.columns]
    
  
  #3.Target Encoding

    if Target_enc and target_col:
        for col in Target_enc:
            if col in train.columns:
                mean_encoding=train.groupby(col)[target_col].mean()
                train[col]=train[col].map(mean_encoding)
            if col in test.columns:
                test[col]=test[col].map(mean_encoding)
                test[col].fillna(train[col].mean(),inplace=True)
    #Now lets see if there is a feature not transformed

    train_check=list(train.select_dtypes(include=["object"]).columns)
    test_check=list(test.select_dtypes(include=["object"]).columns)

    print(f"{train_check},len={len(train_check)}")
    print(f"{test_check},len={len(test_check)}")

    return train,test

                        
```

```python
# Lets decided which encoing method suits for which column
def show_categorical(df):
    for i,column in enumerate(df.select_dtypes(include=["object"]).columns,start=1):
        print(f"{i}. {column} :\t {df[column].unique()}\n")

show_categorical(train_set)
```

```python
# Lets start to improve our model with Feature Engineering

# 1. Lets look at mutual_info
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None)
from sklearn.feature_selection import mutual_info_regression

def make_mi_scores(X, y):
    mi_scores = mutual_info_regression(X, y)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores
```

```python
mi_scores=make_mi_scores(train_set.drop("SalePrice",axis=1),train_set["SalePrice"])
```

```python
# Outlier Detection using Spearman 

def spearman(df,features):
    spr=pd.DataFrame()
    spr["Features"]=features
    spr["Spearman"]=[df[i].corr(df["SalePrice"],"spearman") for i  in features ]
    spr=spr.sort_values("Spearman")

    plt.figure(figsize=(6,0.25*len(features)))
    sns.barplot(data=spr,y="Features",x="Spearman",orient="h",palette=sns.color_palette("deep"))
    plt.title("Spearman Correlation with SalePrice")
    plt.show()
```

```python
## Feature Seleciton
from sklearn.feature_selection import mutual_info_regression,SelectPercentile

real_X_train=train_set.drop("SalePrice",axis=1)
real_y_train=train_set["SalePrice"]

Feature_selection=SelectPercentile(score_func=mutual_info_regression,percentile=20)
X_train_selected=Feature_selection.fit_transform(real_X_train,real_y_train)

selected_features=real_X_train.columns[Feature_selection.get_support()]

X_test=test_set.drop("SalePrice",axis=1)
X_test_selected=X_test[selected_features]

print(f'X_train shape           : {real_X_train.shape}')
print(f'X_train_selected shape  : {X_train_selected.shape}')
print(f'X_test shape            : {X_test.shape}')
print(f'X_test_selected shape   : {X_test_selected.shape}\n')

print(f'Selected Features train :\n{list(selected_features)}')
```

```python
from sklearn.model_selection import train_test_split

X_train_final,X_val,y_train_final,y_val=train_test_split(X_train_selected,real_y_train,test_size=0.2)
print(f"X_train_final shape : {X_train_final.shape}")
print(f"X_val shape         : {X_val.shape}")
print(f"y_train_final shape : {y_train_final.shape}")
print(f"y_val shape         : {y_val.shape}")
```

```python
# Scaling data

from sklearn.preprocessing import StandardScaler,Normalizer

scaler=StandardScaler()

X_train_scaled=scaler.fit_transform(X_train_final)
X_val_scaled=scaler.transform(X_val)

print(f"X_train_scaled shape: {X_train_scaled.shape}")
print(f"X_val_scaled   shape: {X_val_scaled.shape}")
```

```python
X_train_scaled.shape,y_train_final.shape
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from xgboost.sklearn import XGBRegressor



def train_model(model_name):
    if model_name =="Lasso":
        model=Lasso(0.001)
    elif model_name=="Ridge":
        model=Ridge(0.001)
    elif model_name=="ElasticNet":
        model=ElasticNet(0.001)
    elif model_name=="RandomForestRegressor":
        model=RandomForestRegressor(random_state=0)
    elif model_name=="XGBRegressor":
        model=XGBRegressor()
    else:print("Invalid model name")

    model.fit(X_train_scaled,y_train_final)

    print(f'{model_name.capitalize()}  Train Score       : {model.score(X_train_scaled, y_train_final)*100:.2f} %')
    print(f'{model_name.capitalize()}  Validation Score  : {model.score(X_val_scaled , y_val)*100:.2f} %\n')
    
    return model
```

```python

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,root_mean_squared_error

def evaluate_model(y_true,y_pred,data_set="Dataset"):
    mae=mean_absolute_error(y_true,y_pred)
    mse=mean_squared_error(y_true,y_pred)
    rmse=root_mean_squared_error(y_true,y_pred)
    r2=r2_score(y_true,y_pred)

    print(f"model evaluation on {data_set}:")
    print(f"MAE:{mae: .5f} ")
    print(f"MSE:{mse: .5f} ")
    print(f"RMSE:{rmse: .5f} ")
    print(f"R2:{r2: .5f}  ")
```

```python
# Predictions on Lasso

y_train_pred=linear.predict(X_train_scaled)
y_valid_pred=linear.predict(X_val_scaled)

```

```python
# Evaluating
evaluate_model(y_train_final,y_train_pred,"Training Set")
evaluate_model(y_val,y_valid_pred,"VALID Set")
```

```python
from sklearn.model_selection import GridSearchCV

lasso = Lasso()


lasso_parameters = {
    "alpha": np.arange(0.0001, 1.0, 0.01)  
}


lasso_cv = GridSearchCV(
    estimator=lasso, 
    param_grid=lasso_parameters, 
    scoring="neg_root_mean_squared_error",  
    cv=10, 
    n_jobs=-1, 
    verbose=3
)


lasso_cv.fit(X_train_scaled, y_train_final)

# En iyi hiperparametreyi ve skoru yazdır
print(f"Best Alpha: {lasso_cv.best_params_['alpha']}")
print(f"Best RMSE Score: {-lasso_cv.best_score_}")

# En iyi modeli al
best_lasso = lasso_cv.best_estimator_

```
